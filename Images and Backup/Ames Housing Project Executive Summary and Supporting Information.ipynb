# Ames Housing Project Executive Summary 

Ames is a city in central Iowa and is the home of Iowa State University, as well as the home of many US Government agencies. With a population of more than [65,000](https://www.cityofames.org/about-ames/about-ames), it offers a fine mix of small town charm with the amenities more common in larger cities. 

Mathea Stevens Consultant Co. was hired by a local real estate developer (Ames House Flippers Co.) to develop a model to predict housing prices in Ames. With a large swath of data available from the municipality, Ames House Flippers Co. believed there had to be some way to leverage the available data to better inform future capital expenditures. 

### Problem Statement
Develop a model to predict, to the highest degree accuracy available, housing prices in Ames Iowa. 

### Methodology

To create a predictive model, we begin with a thorough investigation of the data provided. There data set contained extensive information of housing features. The data types could be broken down into three disctinct types: *nominal, ordinal, and numeric*. The image to the right breaks down the variables accordingly ([*image courtesy of NYC Data Science Academy*](https://nycdatascience.com/blog/student-works/studying-regression-model-efficacy-on-the-ames-housing-data-set/)). 

<img src = 'https://nycdsa-blog-files.s3.us-east-2.amazonaws.com/2019/09/ML_project_variables-1-1024x576.png' style="width: 600px; float: right;">

Each of these housing features will have to be transformed in different ways in order to be able to create a model. 

The numeric variables, like number of bedrooms and above grade square feet are great indicators for our model, however, variables like MS Subclass (how the municipaility classifies the house) is still a categorical variable.

The Ordinal variables were treated as a ranking scale, with the highest qualities being a 5 and the lowest a 1. 

The Nominal variables, on the other hand, had to ranking to the various categories, and thus were broken down into 1's and 0's. 


From there, the team investigated the high number of missing values, which were represented as NaN in the data set. Through researching reporting methods in Ames, it was determined that the missing values represent the lack of presence of a feature, as opposed to unreported data. These are two very different things. For example, indicating that a house doesn't have a pool with NaN is not the same as saying no one knows if there is a pool present. Therefore, the team was able to replace the missing values with 0's to indicate the lack fo existence for certain features. 

From there, a few additional features were created to help the model identify key relationships. For example, the age of the house at closing was identified through the date built and date sold. Some additional, more complex variables were also created to help strenthen key relationships in the model, such as the squality of finishes per square foot. 

Comparing all variables, existing and engineered, the top correlating features were selected to create our models.

Three different model algorithms were tested to identify which performed best on the training portion of data: Lasso Linear Regression, Ridge Linear Regression and K Nearest Neighbors. Each model performed quite well, allowing us to tweak certain *hyperparameters* to perfect our final model. 


### Conculsions and Recommendations

The model presented can, with 90% confidence, predict the sale price for housing in Ames. 
Key features include:
 - Overall quality
 - Kitchen quality
 - Total liveable square feet squality indicator
 - Whether a house has a concerete foundation
 - Neighborhood

While this list is not exhaustive, it withstands common sense: house prices are often a reflection of the quality of materials used, the quality of the kitchen, square footage, foundation and neighborhood. What this model does tell us, however, is whether a listing over or under priced. 

Through also examining other factors, like garage type, age, fireplace quality, masonry and unfinished basement square footage, the model empowers Ames House Flippers Co. to identify business opportunities, as well as provide some context and additional quantitative measurements to leverage in business decision making.
